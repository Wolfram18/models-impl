Моделирование движения тела (ракеты), брошенного под углом к горизонту - исследование дальности полёта тела (ракеты), брошенного под углом к горизонту при различных значениях начального угла запуска ракеты.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
v0 = 1000 # начальная скорость ракеты (м/с)
g = 9.80665 # ускорение свободного падения (м/с2)
A_grad = np.arange(90, 0, -7.5) # проверяемые углы (град)
A_rad = A_grad*math.pi/180 # (рад)
v0x = v0*np.cos(A_rad) # начальная скорость ракеты по оси x
v0y = v0*np.sin(A_rad) # начальная скорость ракеты по оси y
```

```python
T1 = v0*np.sin(A_rad)/g # время подъёма на максимальную высоту
H = v0**2*np.sin(A_rad)**2/(2*g) # высота подъёма тела
T2 = 2*v0*np.sin(A_rad)/g # время полёта ракеты
L = v0**2*np.sin(2*A_rad)/g # дальность полёта
```

```python
plt.figure(figsize=(10, 5))
plt.title("Координаты полёта ракеты") # заголовок
plt.xlabel("x, км") # ось абсцисс
plt.ylabel("y, км") # ось ординат
plt.grid() # включение отображение сетки

for i in range(len(A_rad)):
    T = np.arange(0, T2[i] + T2[i]/100, T2[i]/100)
    x = v0*T*np.cos(A_rad[i])
    y = v0*T*np.sin(A_rad[i])-(g*T**2)/2
    plt.plot(x/1000, y/1000, label = A_grad[i])

plt.legend(ncol = 3, title = 'Углы')
plt.show()
```

```python
i, = np.where(L == max(L))
print("A = {0:.0f}\N{DEGREE SIGN}".format(float(A_grad[i])))
print("T1 = {0:.2f} с".format(float(T1[i])))
print("H = {0:.2f} км".format(float(H[i]/1000)))
print("T2 = {0:.2f} с".format(float(T2[i])))
print("L = {0:.2f} км".format(float(L[i]/1000)))
```
