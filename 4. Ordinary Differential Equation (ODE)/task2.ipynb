Решение ОДУ второго порядка.

```python
import matplotlib.pyplot as plt # графики
import numpy as np # массивы
from scipy.integrate import odeint # ОДУ
```

```python
# задаём правую часть системы ОДУ
def f(s, t):
    dudt = s[1]
    dvdt = -s[0]
    return [dudt,dvdt]

y0 = [1,0] # начальные значения
t = np.linspace(0,10,100) # массив значений от 0 до 10 (100 точек)
sol = odeint(f, y0, t) # искомая функция u
```

```python
plt.figure(figsize=(6, 6))
plt.title("Решение u''=-cos(t)") # заголовок
plt.xlabel("t") # ось абсцисс
plt.ylabel("u") # ось ординат
plt.grid() # включение отображение сетки
plt.plot(t, sol[:,0], '*b', LineWidth = 2, label = "Найденное") # dudt
plt.plot(t, np.cos(t), 'r', label = "Известное")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.title("Решение вспомогательной u'=-cos(t)") # заголовок
plt.xlabel("t") # ось абсцисс
plt.ylabel("u") # ось ординат
plt.grid() # включение отображение сетки
plt.plot(t, sol[:,1], '*g', LineWidth = 2, label = "Найденное") # dudt
plt.plot(t, -np.sin(t), 'y', label = "Известное")
plt.legend()
plt.show()
```
