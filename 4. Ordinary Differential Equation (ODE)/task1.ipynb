Решение ОДУ первого порядка.

```python
import matplotlib.pyplot as plt # графики
import numpy as np # массивы
from scipy.integrate import odeint # ОДУ
```

```python
# задаём правую часть ОДУ
def f(y,t):
    dydt = 2 * t
    return dydt

y0 = 0 # начальное значение
t = np.linspace(0,20,30) # массив значений от 0 до 20 (30 точек)
y = odeint(f,y0,t) # искомая функция y
```

```python
plt.figure(figsize=(6, 6))
plt.title("Решение y'=2*t") # заголовок
plt.xlabel("t") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.grid() # включение отображение сетки
plt.plot(t, y, '*b', LineWidth = 2, label = "Найденное")
plt.plot(t, t**2, 'r', label = "Известное")
plt.legend()
plt.show()
```
